```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
df1 = pd.read_csv('/content/drive/MyDrive/Datasets/EVStats.csv')
```

```python
df1.replace("Andaman & Nicobar islands", "Andaman and Nicobar Islands", inplace=True)
df1.replace("Jammu & Kashmir", "Jammu and Kashmir", inplace = True)
df1.head()
```

```python
df2 = pd.read_csv('/content/drive/MyDrive/Datasets/Merged_EV.csv')
df2.head()
```

```python
df = pd.merge(left=df2, right=df1, how='left', left_on='State Name', right_on='State')
df.head()
```

```python
df = df.drop('state', axis = 1)
df = df.drop('capital', axis = 1)
df = df.drop('Sl. No', axis = 1)
df = df.drop('State', axis = 1)
df.head()
```

```python
df = df.dropna()
```

```python
#define predictor and response variables
x = df[['No.of Operational Public Charging Stations', 'subsidy', 'road tax', 'petrol', 'diesel']]
y = df['Total in state']

#add constant to predictor variables
x = sm.add_constant(x)

#fit linear regression model
model = sm.OLS(y, x).fit()

#view model summary
print(model.summary())
```

Now we can see that the coefficients of the columns "road tax" and "diesel" have the most correlation with number of EVs and the confidence score of "No.of Operational Public Charging Stations" is less than 0.05. So we use these features to cluster the states and address the EV markets accordingly.

```python
sorted_1 = df.sort_values(by="road tax", ascending=False, kind="mergesort")
```

```python
road_tax1 = sorted_1[:int(len(sorted_1) / 3)]
road_tax1
```

These states have high road tax, so these states are in the infrastructure construction phase and they can be a sink for a new EV market.

```python
sorted_2 = df.sort_values(by="diesel", ascending=True, kind="mergesort")
```

```python
diesel1 = sorted_2[:int(len(sorted_2) / 3)]
diesel1
```

These states have high fuel prices, which bears the adjascent public sentiment of shifting to EVs, so, these can be a new market for EVs.

```python
sorted_3 = df.sort_values(by="No.of Operational Public Charging Stations", ascending=False, kind="mergesort")
```

```python
charging_station1 = sorted_3[:int(len(sorted_3) / 3)]
charging_station1
```

These states have more number of charging stations, and for a new EV startup it will be easier to access the charging stations rather than building the stations from scratch.

```python
x = set(road_tax1['State Name'])
y = set(diesel1['State Name'])
z = set(charging_station1['State Name'])

print(x.intersection(y))
print(y.intersection(z))
print(x.intersection(z))
```

So, it will be better to target the states of Chandigarh, Maharasthra, Karnataka for a entry point for a EV startup
